```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as mcoll
import matplotlib.colors as mcolors
from matplotlib.widgets import Button
import json as js
from sortedcontainers import sortedset

class _Button_callback(object):
    def __init__(self, scenes):
        self.i = 0
        self.scenes = scenes
        self.adding_points = False
        self.added_points = []
        self.adding_lines = False
        self.added_lines = []

    def set_axes(self, ax):
        self.ax = ax
        
    def next(self, event):
        self.i = (self.i + 1) % len(self.scenes)
        self.draw(autoscaling = True)

    def prev(self, event):
        self.i = (self.i - 1) % len(self.scenes)
        self.draw(autoscaling = True)
        
    def add_point(self, event):
        self.adding_points = not self.adding_points
        self.new_line_point = None
        if self.adding_points:
            self.adding_lines = False
            self.added_points.append(PointsCollection())
         
    def add_line(self, event):
        self.adding_lines = not self.adding_lines
        self.new_line_point = None
        if self.adding_lines:
            self.adding_points = False
            self.added_lines.append(LinesCollection())

    def on_click(self, event):
        if event.inaxes != self.ax:
            return
        new_point = (event.xdata, event.ydata)
        if self.adding_points:
            self.added_points[-1].add_points([new_point])
            self.draw(autoscaling = False)
        elif self.adding_lines:
            if self.new_line_point is not None:
                self.added_lines[-1].add([self.new_line_point, new_point])
                self.new_line_point = None
                self.draw(autoscaling = False)
            else:
                self.new_line_point = new_point
        
    def draw(self, autoscaling = True):
        if not autoscaling:
            xlim = self.ax.get_xlim()
            ylim = self.ax.get_ylim()
        self.ax.clear()
        for collection in (self.scenes[self.i].points + self.added_points):
            if len(collection.points) > 0:
                self.ax.scatter(*zip(*(np.array(collection.points))), **collection.kwargs)
        for collection in (self.scenes[self.i].lines + self.added_lines):
            self.ax.add_collection(collection.get_collection())
        self.ax.autoscale(autoscaling)
        if not autoscaling:
            self.ax.set_xlim(xlim)
            self.ax.set_ylim(ylim)
        plt.draw()
```

```python
class Scene:
    def __init__(self, points=[], lines=[]):
        self.points=points
        self.lines=lines

class PointsCollection:
    def __init__(self, points = [], **kwargs):
        self.points = points
        self.kwargs = kwargs
    
    def add_points(self, points):
        self.points = self.points + points

class LinesCollection:
    def __init__(self, lines = [], **kwargs):
        self.lines = lines
        self.kwargs = kwargs
        
    def add(self, line):
        self.lines.append(line)
        
    def get_collection(self):
        return mcoll.LineCollection(self.lines, **self.kwargs)
    
class Plot:
    def __init__(self, scenes = [Scene()], json = None):
        if json is None:
            self.scenes = scenes
        else:
            self.scenes = [Scene([PointsCollection(pointsCol) for pointsCol in scene["points"]], 
                                 [LinesCollection(linesCol) for linesCol in scene["lines"]]) 
                           for scene in js.loads(json)]
        
    def __configure_buttons(self, callback):
        plt.subplots_adjust(bottom=0.2)
        ax_prev = plt.axes([0.6, 0.05, 0.15, 0.075])
        ax_next = plt.axes([0.76, 0.05, 0.15, 0.075])
        ax_add_point = plt.axes([0.44, 0.05, 0.15, 0.075])
        ax_add_line = plt.axes([0.28, 0.05, 0.15, 0.075])
        b_next = Button(ax_next, 'Następny')
        b_next.on_clicked(callback.next)
        b_prev = Button(ax_prev, 'Poprzedni')
        b_prev.on_clicked(callback.prev)
        b_add_point = Button(ax_add_point, 'Dodaj punkt')
        b_add_point.on_clicked(callback.add_point)
        b_add_line = Button(ax_add_line, 'Dodaj linię')
        b_add_line.on_clicked(callback.add_line)
        return [b_prev, b_next, b_add_point, b_add_line]
    
    
    def add_scene(self, scene):
        self.scenes.append(scene)
    
    def add_scenes(self, scenes):
        self.scenes = self.scenes + scenes
        
    def toJson(self):
        return js.dumps([{"points": [np.array(pointCol.points).tolist() for pointCol in scene.points], 
                          "lines":[linesCol.lines for linesCol in scene.lines]} 
                         for scene in self.scenes])
    
    def get_added_points(self):
        if self.callback:
            return self.callback.added_points
        else:
            return None
  
    def get_added_lines(self):
        if self.callback:
            return self.callback.added_lines
        else:
            return None
    
    def get_added_elements(self):
        if self.callback:
            return Scene(self.callback.added_points, self.callback.added_lines)
        else:
            return None
    
    def draw(self):
        plt.close()
        fig = plt.figure()
        self.callback = _Button_callback(self.scenes)
        self.widgets = self.__configure_buttons(self.callback)
        ax = plt.axes(autoscale_on = False)
        self.callback.set_axes(ax)
        fig.canvas.mpl_connect('button_press_event', self.callback.on_click)
        plt.show()
        self.callback.draw()
        
```

```python
import random
def get_random(range_min,range_max):
    return random.uniform(range_min,range_max)

def get_random_from_range(range_min,range_max,n):
    points = []
    for i in range(0,n):
        points.append((get_random(range_min,range_max), get_random(range_min,range_max)))
    return points 

#funkcja generująca odcinki 

def get_random_line(min_range,max_range,amount,epsilon=0.00000001):
    
    segments=[]
    
    for i in range(0,amount):
        points=get_random_from_range(min_range, max_range, 2)
        point1=Point(points[0][0], points[0][1])
        point2=Point(points[1][0], points[1][1])
        if point1.x < point2.x:
            segments.append(Segment(point1,point2))
        if point1.x > point2.x :
            segments.append(Segment(point2,point1))
        segments=reduce(segments, epsilon)
       
    return segments

#funkcja usuwająca odcinki stykające się końcami

def reduce (segments, epsilon=0.00000001):
    for i in segments:
        for j in segments:
            if i!=j:
                if (dist(i.start, j.start) < epsilon or dist(i.end, j.start) < epsilon or dist(i.end, j.end) < epsilon or dist(i.start, j.end) < epsilon) :
                        segments.remove(j)
    return segments
                        
def dist (p1, p2):
      return math.sqrt((p1.x - p2.x) ** 2 + (p1.y - p2.y) ** 2)
```

```python
class Point:    
    def __init__(self, x, y):
        self.x=x
        self.y=y
    
    def to_tuple(self):
        return self.x, self.y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    
    def __hash__(self):
        return (53 + hash(round(float(self.x), 3))) * 53 + hash(round(float(self.y), 3))

# elementy tej klasy przechowywane są s strukturze stanu, zawierają punkt orientacyjny, 
#odcinek wraz z punktem poczatkowym i  koncowym , klucze porównywane wzgl orientacji punktu do innego odcinka   
class StateSegment: 
    def __init__(self,segment):
        self.segment=segment
        self.start=segment.start
        self.end=segment.end
        self.orient_point=segment.start
        
    def to_tuple(self):
        return self.segment.to_tuple() 
    
    def __lt__(self, other):
        if self.orient_point.x >= other.orient_point.x:
            orientation = get_orientation(self.orient_point, other.segment)
        else:
            orientation = -get_orientation(other.orient_point, self.segment)

        if orientation == 1 or \
                       (orientation == 0 and
                        get_orientation(self.end, other.segment) == 1) :
            return True
        else : 
            return False

    def __eq__(self, other):
        if get_orientation(self.orient_point, other.segment) == 0 :
            return True 
        else:
            return False
    
class Segment:
    def __init__(self,start,end):
        self.start=start
        self.end=end
        
    def to_tuple(self):
        return self.start.to_tuple(), self.end.to_tuple()
        
    def __eq__(self, other):
        return self.end == other.end and self.start == other.start

# elementy tej klasy to zdarzenia, porownywane klucze to wartości wpółrzęnej x    
class Event :
    def __init__(self,point,segments,event_type):
        self.segments=segments 
        self.point=point
        self.event_type=event_type
     
    def __lt__(self, other):
        if self.point.x < other.point.x or (self.point.x == other.point.x and self.point.y < other.point.y) : 
            return True
        else : return False
        
    def __gt__(self, other):
        if self.point.x > other.point.x or (self.point.x== other.point.x and self.point.y > other.point.y) : 
            return True
        else : return False
        
    def __eq__(self,other):
        if abs(self.point.x - other.point.x) < 10**-6 and abs(self.point.y - other.point.y) < 10**-6:
            return True
        else:return False

    def __hash__(self, *args, **kwargs):
        return hash(self.point) 

class Intersection(Event): # struktura  przechowuje informacje o przecieciach, w latwy sposob mozna sprawdzic czy dane
                            # przeciecie wystapilo juz wczensiej czy nie
        
    def __init__(self,point,segment1,segment2):
        super().__init__(point,[segment1,segment2], 2)
        self.segment1=segment1
        self.segment2=segment2
       
    def to_tuple(self):
        return self.segment1.to_tuple() + self.segment.to_tuple()
    
    def point_to_tuple():
        return self.point.to_tuple()
    
    def __eq__(self,other):
        return self.segment1==other.segment1 and self.segment2==other.segment2
    
    def __hash__(self, *args, **kwargs):
        return (((13 + hash(self.segment1.start)) * (11 + \
                                                     hash(self.segment1.end))) * (7 + hash(self.segment2.start)) *
                (5 + hash(self.segment2.end)))
```

```python
import math

def factors(p1, p2):
    dy = (p1.y - p2.y)
    dx = (p2.x - p1.x)
    det = (p1.x * p2.y - p2.x * p1.y)
    return dy, dx, -det


def intersection(L1, L2, range1, range2):
    W = L1[0] * L2[1] - L1[1] * L2[0]
    Wx = L1[2] * L2[1] - L1[1] * L2[2]
    Wy = L1[0] * L2[2] - L1[2] * L2[0]
    if W != 0:
        x = Wx / W
        y = Wy / W
        if range1[0] <= x <= range1[1] and range2[0] <= x <= range2[1]:
            return Point(x, y)
        else:
            return None
    else:
        return None
    
def get_above(segment, state):
    index = state.index(segment)
    if index > 0:
        return state[index - 1].segment
    return None

def get_below(segment, state):
    index = state.index(segment)
    if index < len(state) - 1:
        return state[index + 1].segment
    return None  

def get_orientation(point, line,epsilon=0.0001):
    start = line.start
    end = line.end
    if point == start or point == end:
        return 0

    det = start.x * end.y + end.x * point.y + point.x * start.y - end.x * start.y - \
    start.x * point.y - point.x * end.y

    if abs(det) > epsilon:
        if det< 0 : return -1
        elif det > 0 : return 1

    else : return 0


def sweep(data_set):
        
    events=sortedset.SortedSet() # struktura zdarzeń
    for i in data_set:
        events.add(Event(i.start,[i], "start"))
        events.add(Event(i.end, [i], "end"))
        
    state = sortedset.SortedList() # struktura stanu

    while len(events):
        event = events.pop(0)
        if event.event_type == "start":

            st = StateSegment(event.segments[0])
            find_intersections(state, st, events)
        elif event.event_type == "end":

            st = StateSegment(event.segments[0])
            above_neighbour = get_above(st, state)
            below_neighbour = get_below(st, state)

            state.remove(st)
            add_intersection(above_neighbour, below_neighbour, 1, events)
        else:
            above_segment = StateSegment(event.segments[0])
            below_segment = StateSegment(event.segments[1])

            state.remove(above_segment)
            state.remove(below_segment)

            above_segment.orient_point = event.point
            below_segment.orient_point = event.point

            find_intersections(state, above_segment, events)

            find_intersections(state, below_segment, events)
    
    
    print(len(intersections))
    for i in intersections:
        print("intersection in ",i.point.to_tuple(), "between ",i.segment1.to_tuple(), "and ", i.segment2.to_tuple())
    return intersections      
    

def intersection_point(segment1, segment2, first_orient):
    orientation = get_orientation(segment1.end, segment2)
    if orientation != first_orient:
        line1 = factors(segment1.start, segment1.end)
        line2 = factors(segment2.start, segment2.end)
        point = intersection(line1, line2, (segment1.start.x, segment1.end.x), (segment2.start.x, segment2.end.x))
        if point:
            return point
        return None
    return None


def add_intersection(segment, neighbour, orientation, events):
    if neighbour and segment:
        point = intersection_point(segment, neighbour, orientation)
        if point:
            inter = Intersection(point, segment, neighbour)
            if inter not in intersections:
                intersections.add(inter)
                events.add(Event(point, [segment, neighbour], "intersection"))


def find_intersections(state, st, events):
    state.add(st)
    above_neighbour = get_above(st, state)
    below_neighbour = get_below(st, state)
    add_intersection(st.segment, above_neighbour, -1, events)
    add_intersection(st.segment, below_neighbour, 1, events)
```

```python
def Visualization(data):

    result=sweep(data)
    steps = []
    lines = []
    for i in range(0,len(data)):
        steps.append(data[i].start.to_tuple())
        steps.append(data[i].end.to_tuple())
        lines.append(data[i].to_tuple())
        
    for i in result:
        steps.append(i.point.to_tuple())
  
    steps=sorted(steps,key=lambda s : (s[0],s[1]))
    result = [res.point.to_tuple() for res in result]
    y_max=steps[0][1]
    y_min=steps[0][1]
    for i in range(0,len(steps)):
        y_max=max(y_max, steps[i][1])
        y_min=min(y_min,steps[i][1])
    scenes=[]
    inter = []
    for i in range(0,len(steps)):
        if steps[i] in result:
            inter.append(steps[i])

        scenes.append(Scene([ PointsCollection(list(inter), color="red")], [LinesCollection(lines),
                                                        LinesCollection([[(steps[i][0],y_max), (steps[i][0],y_min)]], \
                                                                        color="orange")]))
                
    return scenes
    
```
